**Goal:** To understand how size affects house price.
**Algorithm:** Linear Regression (Supervised Learning)

***üìí Create Small House Price Dataset***

```python
import pandas as pd
import numpy as np

data = {
    "Size_sqft": [800, 1000, 1200, 1500, 1800],
    "Bedrooms": [2, 2, 3, 3, 4],
    "Price_in_lakhs": [40, 50, 65, 80, 95]
}

# Create DataFrame to store this "data" into Table
house_df = pd.DataFrame(data)
print(house_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(7,4))

plt.title("House Price based on Size",color="red")

plt.xlabel("House Price",color="red")

plt.ylabel("House Size",color="red")

sns.scatterplot(x=house_df["Price_in_lakhs"], y=house_df["Size_sqft"],data=house_df)

plt.show()
```

**üëâ Now Create Models on my house_df**

```python
# Import from scikit learn library
from sklearn.linear_model import LinearRegression

# for supervised learning:- we have 'X' which represents the input features and 'Y' which represents the output label.
y = house_df["Price_in_lakhs"]
X = house_df[["Size_sqft"]]

model = LinearRegression()

# Now train the model
model.fit(X,y)

# Now predict house price for 1500 room_size
room_size_per_sqr = [[1600]]
prediction = model.predict(room_size_per_sqr)
print(f"Predicted Price in lakh for 1600 per square room is {prediction}")
```
